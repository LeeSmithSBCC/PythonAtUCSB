### Tonight we will continue our journey with LISTS and LOOPS
#### In our Book:  Chapters 4 and 6

```python
# Remember how we created simple lists
leeList = [1,2,3,'Lee', 'UCSB', 42.0, 66.0, 101.0]
print(leeList)
```

```python
# ... and how to access elements of the list
#### Remember that Python Containers are ZERO based
print(leeList[2])
```

```python
# ... or part of the list
print(leeList[1:4])
```

### OK, but can we quickly print out a subset of elements?
#### That requires learning about simple loops

### This is process called iteration using the keyword 'for'

```python
for x in leeList:
    print(x)
```

### Remember how to run elements on a single line?

```python
for x in leeList:
    print(x, end = " ")
```

### just because I am using x, we are NOT doing MATH
#### x is just a placeholder ...

```python
for apple in leeList:
    print(apple, end=" ")
```

### Strings are also lists of characters!

```python
leeName = 'lee'
for eagle in leeName:
    print(eagle, end=" ")
```

### This is a list comprehension, we will use more later

```python
[print(i, end = " ") for i in leeList];
```

### Also you can create a sequence of elements

```python
for i in range(10):
    print(i, end=' ')
```

### Finally, here is a cool new way to create sequences is numpy
#### Notice that this creates a new LIST !

```python
import numpy as np
y = np.arange(10)
print(y, end = ' ')
```

### OK, now back to creating business graphics:  BAR GRAPHS

```python
from matplotlib.pyplot import *
import numpy as np
 
objects = ('Python', 'C++', 'Java', 'Perl', 'Scala', 'Lisp')
colors = ['red', 'yellow', 'green', 'blue', 'orange','black']
x = np.arange(len(objects))
heights = [10,8,6,4,2,1]
bar(x, heights, align='center', alpha=0.5, color=colors)
xticks(x, objects)
ylabel('Usage')
title('Programming language usage')
 
show();
```
