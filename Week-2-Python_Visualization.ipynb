```python
### NOAA 1832 to 2014 CO2 measurements
# __CO2_NOAA_1959_2014_BEST.py

from pylab import *

x = []
y = []

readFile = open('NOAA_1832-1978.txt','r')
sepFile = readFile.read().strip().replace('\t','').splitlines()
readFile.close()

for plotPair in sepFile:
    xAndY = plotPair.split()
    x.append(float(xAndY[0]))
    y.append(float(xAndY[1]))

figure(1, figsize=(12, 5.5))
plot(x,y,'g+-', markersize=20, label='Measured Levels')
axhline(280, color = 'red', linewidth=5,label='440k year average')
axhline(400, color = 'red', linewidth=3)
ylim(250,450)
title(r'$CO_{2}$ Levels Since 1832', fontsize=26)
xlabel('Time AD (years)', fontsize=20, weight='bold')
ylabel(r'$CO_{2}$ (ppmv)', fontsize=20, weight='bold')
xticks(fontsize=16)
yticks(fontsize=16)
legend()

show()
```

```python
### NOAA 1832 to 2014 CO2 measurements
# __CO2_NOAA_1959_2014_BEST_Pandas.py

from pylab import *
import pandas as pd

leeGlobalWarming = pd.read_csv('NOAA_1832-1978_Panedas.txt', sep=' ',header=None)

figure(1, figsize=(12, 5.5))
plot(leeGlobalWarming[0],leeGlobalWarming[1],'g+-', markersize=20, label='Measured Levels')
axhline(280, color = 'red', linewidth=5,label='440k year average')
axhline(400, color = 'red', linewidth=3)
ylim(250,450)
title(r'$CO_{2}$ Levels Since 1832', fontsize=26)
xlabel('Time AD (years)', fontsize=20, weight='bold')
ylabel(r'$CO_{2}$ (ppmv)', fontsize=20, weight='bold')
xticks(fontsize=16)
yticks(fontsize=16)
legend()

show()
```
